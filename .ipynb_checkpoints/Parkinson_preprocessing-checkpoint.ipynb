```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

```python
park_df= pd.read_csv(os.path.join("Resources/Parkinson_disease.csv"))
park_df.head()
```

```python
park_df.shape
```

```python
park_df.info()
```

```python
# check status column
park_df["status"].value_counts()

# We can see that this is clearly unbalanced, need to fix 
```

```python
# check null values, there are no null values
park_df[park_df.isnull().any(axis=1)]
```

```python
# Drop Name column, serves no purpose for Model
park_df= park_df.drop("name",axis=1)
park_df.head()
```

```python
# Plot Unbalanced Classes
sns.countplot(x="status", data=park_df)
```

```python
# Box Plots for Data between 1 and 0

k=[]
for i in park_df.columns:
    for j in park_df[i]:
        if (j<1 and j>0):
            k.append(i)
            break
            
fig, ax = plt.subplots(figsize=(20,10))
park_df.boxplot(k,ax=ax)
```

```python
# Flip Data to Read it all
park_df.describe().T
```

### Analysis From Above
* In general, St. Dev is close to one another for the most part. This means data is near the Meann of dataset.
    If we have a high standard deviation, we will have data that is far from its mean.
